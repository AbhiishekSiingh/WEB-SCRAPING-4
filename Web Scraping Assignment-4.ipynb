# Q1) Scrape the details of most viewed videos on YouTube from Wikipedia.

# Q8. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/ You have to find the following details:
        
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute 
G) Year

Note: - from the home page you have to go to the Show All Dataset page through code.

```python
driver=webdriver.Chrome()
```

```python
driver.get(' https://archive.ics.uci.edu/')
```

```python
dataset=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
dataset.click()
```

```python
Task=[]
year=[]
df['Column'] = df['Column'].apply(lambda x: x if x is not None else '-')
```

```python
Dataset_name=[]
Data_type=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
```

```python
import time
start=0
end=65
for page in range(start,end):
    name_tags=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')[:500]
    for i in name_tags:
        Dataset_name.append(i.text)
        
    data_type=driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"]/span')[0:550:4]
    for i in data_type:
        Data_type.append(i.text)
        
    attributetype_tags=driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"]/span')[1:550:4]
    for i in attributetype_tags:
        Attribute_type.append(i.text)
        
    instance_tags=driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"]/span')[2:550:4]
    for i in instance_tags:
        No_of_instances.append(i.text)
        
    attribute_tags=driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"]/span')[3:550:4]
    for i in attribute_tags:
        No_of_attribute.append(i.text)
        
    next_button=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
    next_button.click()
    time.sleep(2)
```

```python
print(len(Dataset_name),len(Data_type),len(Attribute_type),len(No_of_instances),len(No_of_attribute))
```

```python
df1=pd.DataFrame(Dataset_name)[:500]
df2=pd.DataFrame(Data_type)[:500]
df3=pd.DataFrame(Attribute_type)[:500]
df4=pd.DataFrame(No_of_instances)[:500]
df5=pd.DataFrame(No_of_attribute)[:500]
```

```python
result = pd.concat([df1,df2,df3,df4,df5], axis=1)
result
```
